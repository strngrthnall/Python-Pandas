## Criando novas Variaveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('data/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
dados['Valor m²'] = dados['Valor']/dados['Area']
dados.head(10)
```

#### Arredondando com duas casas deciamais

```python
dados['Valor m²'] = dados['Valor m²'].round(2)
dados.head(10)
```

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto']/dados['Area']).round(2)
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

## Excluindo Variaveis

```python
dados_aux = pd.DataFrame(dados[['Valor Bruto', 'Valor m²', 'Valor Bruto m²', 'Tipo Agregado']])
dados_aux.head(10)
```
